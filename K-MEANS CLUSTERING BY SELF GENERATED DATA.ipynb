# Each of the 100 rows has been assigned to a specific centroid, resulting in a total of 100 clusters.

```python

class KMeans:
    def __init__(self,n_clusters=2, max_iter=100): 
        self.n_clusters=n_clusters
        self.max_iter= max_iter
        self.centriod = None 
        
        
    def fit_predict(self,X):
        random_index=random.sample(range(0,X.shape[0]),self.n_clusters)
        self.centriod=X[random_index]
        
        for i in range(self.max_iter):
            #assign cluster
            cluster_group = self.assign_cluster(X)
        
            #move centriod
            old_centriods =  self.centriod
            self.centriod= self.move_centriod(X,cluster_group)
            
            #check finish
            if (old_centriods == self.centriod).all:
                break
                
                
        return cluster_group # returning the last , final cluster after iterations        
            
     #assign cluster  
    
    def assign_cluster(self,X):
        cluster_group= []
        distances= []
        
        
        for row in X:
            for centriod in self.centriod:
                distances.append(np.sqrt(np.dot(row-centriod, row-centriod)))
            min_distance= min(distances )
            index_pos= distances.index(min_distance)
            cluster_group.append(index_pos)
            distances.clear()
        return np.array(cluster_group)
    
    
    #move centriod
    
    def move_centriod(self, X,cluster_group):
        new_centriod =[]
        cluster_type = np.unique(cluster_group)
        
        
        for type in cluster_type:
            new_centriod.append(X[cluster_group==type].mean(axis=0))
            
        return np.array(new_centriod)     
            
```

# The clustering process involves the following steps:

1- Decide Clusters:

Determine the number of clusters either through the Within-Cluster Sum of Squares (WCSS) method or by a predefined criterion, reflecting the desired cluster count.


2- Select Random Centroids:

Randomly choose initial centroids, where the number of centroids matches the previously decided number of clusters. This initialization sets the starting points for the clustering process.


3-Assign Clusters:

Associate each data point with the nearest centroid based on the Euclidean distance. The point is assigned to the cluster represented by the centroid with the minimum distance.


4- Move Centroids:

Recalculate the centroids by computing the mean of the data points belonging to each cluster. This step iterates until the updated centroids stabilize and there is no significant change from the previous iteration.


5-Check Finish:

Conclude the clustering process when the centroids no longer change between consecutive iterations, signifying that the algorithm has converged and the clusters are established.


# In summary, this iterative process refines the cluster assignments and centroid positions until a stable clustering solution is reached, demonstrating the power of clustering in organizing and grouping data.
